**This code is for testing how good the API's are working**

imports:

```python
import requests

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import io
from PIL import Image
import urllib.parse
```

set the base url, you can set it to local url if you are running this locally or use the `https://modifinder.gnps2.org/` for the version running on gnps

```python
base_url = "http://localhost:5002" # change this to the URL of your local server or https://modifinder.gnps2.org/
```

## Draw Molecule

```python
url = base_url + "/api/visualizer/draw_molecule?mol=CCO.png"
print(url)
response = requests.get(url)
img = response.content
# img was created the following way:
            # img = Image.fromarray(result)
            # buffer = BytesIO()
            # img.save(buffer, format='PNG')
            # buffer.seek(0)
# to display the image in the notebook, we need to convert it to a numpy array
img = Image.open(io.BytesIO(img))
img = np.array(img)
plt.imshow(img)
plt.axis('off')
plt.show()
```

## Draw Modification

```python
url = base_url + "/api/visualizer/draw_modifications?mol1=CCO&mol2=COO.png"
print(url)
response = requests.get(url)
img = response.content
img = Image.open(io.BytesIO(img))
img = np.array(img)
plt.imshow(img)
plt.axis('off')
plt.show()
```
